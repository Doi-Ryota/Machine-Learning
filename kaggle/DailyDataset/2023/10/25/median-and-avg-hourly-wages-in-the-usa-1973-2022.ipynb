```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.simplefilter("ignore")

pd.set_option("display.max_columns",None)
```

# Loading Data

```python
df = pd.read_csv("/kaggle/input/median-and-avg-hourly-wages-in-the-usa-1973-2022/median_average_wages.csv")
print(df.shape)
df.head(3)
```

# Data Observation

```python
df.info()
```

```python
sns.lineplot(x=df["year"],y=df["median"],label='median')
sns.lineplot(x=df["year"],y=df["average"],label='average')
plt.legend()
```

```python
sns.lineplot(x=df["year"],y=df["average"]-df["median"])
```

## A large difference between the mean and median indicates that the distribution of salary data is asymmetric. It indicates that the distribution of salary data has been highly skewed in recent years.

```python
sns.lineplot(x=df["year"],y=df["men_median"],label='median')
sns.lineplot(x=df["year"],y=df["men_average"],label='average')
plt.legend()
```

```python
sns.lineplot(x=df["year"],y=df["men_average"]-df["men_median"])
```

```python
sns.lineplot(x=df["year"],y=df["women_median"],label='median')
sns.lineplot(x=df["year"],y=df["women_average"],label='average')
plt.legend()
```

```python
sns.lineplot(x=df["year"],y=df["women_average"]-df["women_median"])
```

## The gap between rich and poor is greater for men than for women.

```python
sns.lineplot(x=df["year"],y=df["men_median"]-df["women_median"],label='median')
sns.lineplot(x=df["year"],y=df["men_average"]-df["women_average"],label='average')
plt.legend()
```

## Look at the income gap between men and women.
## The median difference is smaller, but the mean difference is larger.
## This could be interpreted in a variety of ways. We would also like to know the composition of the number of people.

```python
sns.lineplot(x=df["year"],y=df["white_median"],label='median')
sns.lineplot(x=df["year"],y=df["white_average"],label='average')
plt.legend()
```

```python
sns.lineplot(x=df["year"],y=df["white_average"]-df["white_median"])
```

```python
sns.lineplot(x=df["year"],y=df["black_median"],label='median')
sns.lineplot(x=df["year"],y=df["black_average"],label='average')
plt.legend()
```

```python
sns.lineplot(x=df["year"],y=df["black_average"]-df["black_median"])
```

```python
sns.lineplot(x=df["year"],y=df["hispanic_median"],label='median')
sns.lineplot(x=df["year"],y=df["hispanic_average"],label='average')
plt.legend()
```

```python
sns.lineplot(x=df["year"],y=df["hispanic_average"]-df["hispanic_median"])
```

## Whites have a growing gap between rich and poor.

```python
sns.lineplot(x=df["year"],y=df["white_median"]-df["black_median"],label='median')
sns.lineplot(x=df["year"],y=df["white_average"]-df["black_average"],label='average')
plt.title("white vs black")
plt.legend()
```

```python
sns.lineplot(x=df["year"],y=df["white_median"]-df["hispanic_median"],label='median')
sns.lineplot(x=df["year"],y=df["white_average"]-df["hispanic_average"],label='average')
plt.title("white vs hispanic")
plt.legend()
```

```python
sns.lineplot(x=df["year"],y=df["black_median"]-df["hispanic_median"],label='median')
sns.lineplot(x=df["year"],y=df["black_average"]-df["hispanic_average"],label='average')
plt.title("black vs hispanic")
plt.legend()
```

## Hispanics and blacks have almost the same difference in mean and median. There is little disparity.
## The gap between whites and other people is widening.

## Hypothesis that the number of wealthy white people is increasing
