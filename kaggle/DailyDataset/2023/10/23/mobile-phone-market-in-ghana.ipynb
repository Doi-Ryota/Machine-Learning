```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.simplefilter("ignore")

pd.set_option("display.max_columns",None)
```

## Loading Data

```python
df = pd.read_csv("/kaggle/input/mobile-phone-market-in-ghana/Mobile-Phones.csv")
print(df.shape)
df.head(3)
```

## Data Observation

### brand

```python
df["brand"].value_counts()
```

```python
tmp_df = df.groupby("brand").agg({"price(¢)":["min","max","mean","std"]})
tmp_df
```

```python
mean_prices = tmp_df["price(¢)"]["mean"]
brands = mean_prices.index
plt.figure(figsize=(10, 6))
sns.barplot(x=brands, y=mean_prices)
```
